The MiraBest Batched Dataset is comprised of a total of 1256 images, with their class denoted by a three-digit number.

First digit: FRI (1), FRII (2), hybrid source (3). Second digit: confidently classified (0), uncertaintly classified (1). Third digit: standard morphology (0), double-double (1), wide-angle tail (2), head-tail (4).

Images were obtained from the VLA FIRST sky survey via SkyView virtual telescope. All images have dimensions of 150 x 150 pixels, with one pixel equivalent to an angular size of 1.8". The dataset is split into 7 training batches and a test batch, each composed of 157 images. The number of objects of each class in each batch is roughly consistent between batches, and the test batch contains at least one example of every class.

Source: https://zenodo.org/record/4288837#.Y16QTHbP2Uk

Not very informative. Need to process the image

```python
mx_val = np.max(abs(image))
```

```python
img_norm = abs(image/mx_val)
plt.imshow(img_norm)
```

A bit better but still not very sufficient (Although now it barely does look like a quasar compared to the first attempt)

```python
batch = dataset_mlsst.take(9)
```

```python
fig = plt.figure(figsize=(16,12))
i=1

for sample in batch:

    image, label = sample["image"], sample["label"]

    mx_val = np.max(abs(image))
    img_norm = abs(image/mx_val)

    ax = fig.add_subplot(3,3,i)
    ax.imshow(img_norm)
    ax.set_xlabel(label.numpy())
    ax.set_xticks([])
    ax.set_yticks([])

    i+=1

plt.tight_layout
plt.show()
```

Since the datasets main focus is to examine "Robustness of Deep Learning Models for Galaxy Morphology Classification", the labels for the classification are likely to be the following:

0) Elliptical and Spheroidal Galaxies.
1) S0 (lenticular) and Spiral Galaxies.
2) Irregular Galaxies.
